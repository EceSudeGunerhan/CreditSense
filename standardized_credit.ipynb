```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('output.csv')

# Bağımsız ve hedef değişkenleri ayırma
X = data.drop(columns=['BAD'])  # Bağımsız değişkenler
y = data['BAD']  # Hedef değişken

# StandardScaler ile standartlaştırma
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Standartlaştırılmış veriyi DataFrame olarak oluşturma
standardized_data = pd.DataFrame(X_standardized, columns=X.columns)
standardized_data['BAD'] = y  # Hedef değişkeni ekleme

# Standartlaştırılmış veriyi CSV olarak kaydetme
standardized_data.to_csv('standardized_output.csv', index=False)
```
